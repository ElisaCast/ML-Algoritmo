```python
# Valores atipicos en DURATION
duration_outliers = total_data["duration"].describe()
print(duration_outliers)

# IQR (rango intercuartil)
duration_iqr = duration_outliers["75%"] - duration_outliers["25%"]
upper_limit = duration_outliers["75%"] + 1.5 * duration_outliers
lower_limit = duration_outliers["25%"] - 1.5 * duration_outliers
print(f"Los limites superiose inferios son {round(upper_limit, 2)} y {round(lower_limit, 2)}, y un rango intercuartil de {round(duration_iqr, 2)}")

# Borrar valores atipicos
total_data = total_data[total_data["duration"] > 100]
```

```python
# Valores atipicos en PREVIOUS
duration_outliers = total_data["previous"].describe()
print(duration_outliers)

# IQR (rango intercuartil)
duration_iqr = duration_outliers["75%"] - duration_outliers["25%"]
upper_limit = duration_outliers["75%"] + 1.5 * duration_outliers
lower_limit = duration_outliers["25%"] - 1.5 * duration_outliers
print(f"Los limites superiose inferios son {round(upper_limit, 2)} y {round(lower_limit, 2)}, y un rango intercuartil de {round(duration_iqr, 2)}")

# Borrar valores atipicos
total_data = total_data[total_data["previous"] > 0]
```

```python
# Escalado de caracteristicas MinMax

from sklearn.preprocessing import MinMaxScaler

total_data["job_n"] = pd.factorize(total_data["job"])[0]
total_data["marital_n"] = pd.factorize(total_data["marital"])[0]
total_data["education_n"] = pd.factorize(total_data["education"])[0]
total_data["default_n"] = pd.factorize(total_data["default"])[0]
total_data["housing_n"] = pd.factorize(total_data["housing"])[0]
total_data["loan_n"] = pd.factorize(total_data["loan"])[0]
total_data["contact_n"] = pd.factorize(total_data["contact"])[0]
total_data["month_n"] = pd.factorize(total_data["month"])[0]
total_data["day_of_week_n"] = pd.factorize(total_data["day_of_week"])[0]
total_data["poutcome_n"] = pd.factorize(total_data["poutcome"])[0]
total_data["y_n"] = pd.factorize(total_data["y"])[0]
num_variables = ["job_n", "marital_n", "education_n", "default_n", "housing_n", "loan_n", "contact_n", "month_n", "day_of_week_n", "poutcome_n",
                 "age", "duration", "campaign", "pdays", "previous", "emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed", "y_n"]

scaler = MinMaxScaler()
scal_features = scaler.fit_transform(total_data[num_variables])
total_data_scal = pd.DataFrame(scal_features, index = total_data.index, columns = num_variables)
total_data_scal.head()
```

```python
# Seleccion de caracteristicas

from sklearn.feature_selection import chi2, SelectKBest
from sklearn.model_selection import train_test_split

X = total_data_scal.drop("y_n", axis = 1)
y = total_data_scal["y_n"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)

selection_model = SelectKBest(chi2, k = 5)
selection_model.fit(X_train, y_train)
ix = selection_model.get_support()
X_train_sel = pd.DataFrame(selection_model.transform(X_train), columns = X_train.columns.values[ix])
X_test_sel = pd.DataFrame(selection_model.transform(X_test), columns = X_test.columns.values[ix])

X_train_sel.head()
```

```python
X_train_sel["y_n"] = list(y_train)
X_test_sel["y_n"] = list(y_test)
X_train_sel.to_csv("../data/processed/clean_train.csv", index = False)
X_test_sel.to_csv("../data/processed/clean_test.csv", index = False)
```

```python
train_data = pd.read_csv("../data/processed/clean_train.csv")
test_data = pd.read_csv("../data/processed/clean_test.csv")

train_data.head()
```

```python
X_train = train_data.drop(["y_n"], axis = 1)
y_train = train_data["y_n"]
X_test = test_data.drop(["y_n"], axis = 1)
y_test = test_data["y_n"]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

bank_cm = confusion_matrix(y_test, y_pred)

# Dibujaremos esta matriz para hacerla mÃ¡s visual
cm_df = pd.DataFrame(bank_cm)

plt.figure(figsize = (3, 3))
sns.heatmap(cm_df, annot=True, fmt="d", cbar=False)

plt.tight_layout()

plt.show()
```

La interpretaciÃ³n de una matriz de confusiÃ³n es la siguiente:

- **Verdaderos positivos** (*TP, True positive*): Se corresponde con el nÃºmero `81` y son los casos en los que el modelo predijo positivo (no supervivencia) y la clase real tambiÃ©n es positiva.
- **Verdaderos negativos** (*TN, False negative*): Se corresponde con el nÃºmero `53` y son los casos en los que el modelo predijo negativo (supervivencia) y la clase real tambiÃ©n es negativa.
- **Falsos positivos** (*FP, False positive*): Se corresponde con el nÃºmero `15` y son los casos en los que el modelo predijo positivo y la clase real es negativa.
- **Falsos negativos** (*FN, False negative*): Se corresponde con el nÃºmero `15` y son los casos en los que el modelo predijo negativo y la clase real es positiva.

Estas cuatro medidas se utilizan a menudo para calcular mÃ©tricas mÃ¡s complejas.

```python
# Busqueda ALEATORIA
import numpy as np
from sklearn.model_selection import RandomizedSearchCV

# Definimos los parÃ¡metros que queremos ajustar
hyperparams = {
    "C": np.logspace(-4, 4, 20),
    "penalty": ["l1", "l2", "elasticnet", None],
    "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}

# Inicializamos la bÃºsqueda aleatoria
random_search = RandomizedSearchCV(model, hyperparams, n_iter = 100, scoring = "accuracy", cv = 5, random_state = 42)
random_search
```
